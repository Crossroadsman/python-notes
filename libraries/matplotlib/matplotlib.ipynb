```python
import matplotlib.pyplot as plt
```

```python
line1 = {
    'data': {
        'x_vals': [1, 2, 3, 4],
        'y_vals': [1, 4, 9, 16],
    },
    'color': 'green',
    'linestyle': 'dashdot',
    'label': 'dashdot',
}
line2 = {
    'data': {
        'x_vals': [2, 3, 4, 5],
        'y_vals': [3, 1, 12, 8],
    },
    'color': '#2B5B84',
    'linestyle': 'dashed',
    'label': 'dashed',
}

y_label = "Important figures"

plt.plot('x_vals', 'y_vals', **line1)
plt.plot('x_vals', 'y_vals', **line2)
plt.ylabel(y_label)
plt.legend()
plt.show()
```

```python
lims = {'x': [0, 6], 'y': [0, 20],}

# create first panel (rows, cols, panel#)
panel_1 = plt.subplot(2, 1, 1)
plt.plot('x_vals', 'y_vals', **line1)
panel_1.set_xlim(lims['x'])
panel_1.set_ylim(lims['y'])

# create second panel
panel_2 = plt.subplot(2, 1, 2)
plt.plot('x_vals', 'y_vals', **line2)
panel_2.set_xlim(lims['x'])
panel_2.set_ylim(lims['y'])

plt.show()
```
