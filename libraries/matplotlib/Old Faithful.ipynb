```python
import csv
```

```python
INPUT_FILE = 'old_faithful.csv'
```

```python
with open(INPUT_FILE, 'r') as of_data:
    of = list(csv.reader(of_data))
```

```python
# Take a look at the shape of the data
for row in of:
    print(row)
```

```python
# Since the data include a header row, we need to remove it
of.pop(0)
```

```python
import matplotlib.pyplot as plt
```

```python
# plt settings
plt.figure(figsize=(7.5, 4.25))
plt.style.use('classic')
```

```python
eruptions = []  # durations
waitings = []  # time between eruptions
for row in of:
    eruptions.append(float(row[0]))
    waitings.append(float(row[1]))
```

```python
HIST_BINS = 8

# (rows, columns, plot number)
plt.subplot(3, 2, 1)
plt.boxplot(eruptions)
plt.title('Old Faithful Eruptions')
plt.xticks([1], ['Eruptions'])
plt.xlabel('Eruption duration (minutes)')

plt.subplot(3, 2, 2)
plt.hist(eruptions, HIST_BINS)
plt.title('Old Faithful Eruptions')
plt.xlabel('Eruption duration (minutes)')

plt.subplot(3, 2, 3)
plt.boxplot(waitings)
plt.title('Intervals Between Old Faithful Eruptions')
plt.xticks([1], ['Wait Times'])
plt.xlabel('Wait Times (minutes)')

plt.subplot(3, 2, 4)
plt.hist(waitings, HIST_BINS)
plt.title('Intervals Between Old Faithful Eruptions')
plt.xlabel('Wait Times (minutes)')

plt.subplot(3, 2, 5)
plt.scatter(eruptions, waitings)
plt.title('Old Faithful Eruptions')
plt.xlabel('Eruption duration (minutes)')
plt.ylabel('Wait Times (minutes)')

# adjust padding
plt.tight_layout()

plt.show()
```
